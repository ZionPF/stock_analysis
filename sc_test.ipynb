```python
from word_segmentation import stock_code_dict

data_path = '/data/jupyter/stock/data/0520_1.csv'

dict_code = stock_code_dict()
```

```python

with open(data_path, 'r', encoding='utf-8') as data:
    for line in data.readlines():
        list_line = line.strip().split(',')
        if list_line[0].replace('"', '') not in dict_code:
            print(list_line[0] + ',' + list_line[1] + ',' + list_line[2])
```

```python
file_path = '/data/jupyter/stock/data/plate.csv'
plate_list = []
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file.readlines():
        plate_list.append(line.replace('\n', '').replace('\r', ''))
num_list = []
for i in range(60):
    num_list.append(i)

plate_dict = dict(zip(plate_list, num_list))
print(plate_dict)
```

```python
path_whatever = '/data/jupyter/stock/data/0520_2.csv'
with open(data_path, 'r', encoding='utf-8') as data:
    with open(path_whatever, 'w', encoding = 'utf-8') as whatever:
        for line in data.readlines():
            list_line = line.strip().split(',')
            if str(list_line[2].replace('"', '')) not in plate_dict:
                if str(list_line[2].replace('"', '') + '行业') in plate_dict:
                    print(list_line[0] + ',' + list_line[1] + ',' + list_line[2])
                    line = line.replace(list_line[2].replace('"', ''), list_line[2].replace('"', '') + '行业')
                    whatever.writelines(line)
                else:
                    whatever.writelines(line)
            else:
                whatever.writelines(line)
```
